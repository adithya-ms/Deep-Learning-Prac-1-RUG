```python
import numpy as np
import pandas as pd
from time import time
import tensorflow.keras
from tensorflow.keras.models import Sequential
import tensorflow.keras.layers as layers
from keras.utils.np_utils import to_categorical
from tensorflow.keras.callbacks import TensorBoard
import os
import time
import datetime
```

```python
from keras.datasets import cifar10
```

```python
batch_size = 32
num_classes = 10
epochs = 100
#data_augmentation = False
num_predictions = 20
save_dir = os.path.join(os.getcwd(), 'saved_models')
```

```python
(x_train,y_train),(x_test,y_test) = cifar10.load_data()
print('x_train shape:',x_train.shape)
```

```python
y_train = tensorflow.keras.utils.to_categorical(y_train,num_classes)
y_test = tensorflow.keras.utils.to_categorical(y_test,num_classes)
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
```

## Activation Layer : ReLU

```python
relumodel = Sequential()

relumodel.add(layers.Conv2D(filters=6, kernel_size=(5, 5), activation='relu', input_shape=(32,32,3)))
relumodel.add(layers.AveragePooling2D())

relumodel.add(layers.Conv2D(filters=16, kernel_size=(3, 3), activation='relu'))
relumodel.add(layers.AveragePooling2D())

relumodel.add(layers.Flatten())

relumodel.add(layers.Dense(units=120, activation='relu'))

relumodel.add(layers.Dense(units=84, activation='relu'))

relumodel.add(layers.Dense(units=10, activation = 'softmax'))
relu_model_name = 'lenet_relu_cifar10_trained_model.h5'

NAME_lenet = "Lenet-ReLU-{}".format(int(time.time()))

tensor_board_1 = TensorBoard(log_dir = os.path.join("logs","LeNet_ReLU",
                                                  "fit",
                                                  datetime.datetime.now().strftime("%Y%m%d-%H%M%S"),
                                                 ))

#Compile the model
relumodel.compile(loss='categorical_crossentropy',
              optimizer='sgd',
              metrics=['accuracy'])

#Training starts
relumodel.fit(x_train, y_train,
              callbacks=[tensor_board_1],
              batch_size=batch_size,epochs=epochs,validation_data=(x_test, y_test),shuffle=True)

if not os.path.isdir(save_dir):
    os.makedirs(save_dir)
relu_model_path = os.path.join(save_dir, relu_model_name)
relumodel.save(relu_model_path)
```
